# задание 1
    Определите, какому фильму было выставлено больше всего оценок 5.0.



```python

movies = pd.read_csv('1_home_work/task_1/movies.csv')
ratings = pd.read_csv('1_home_work/task_1/ratings.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
ratings['rating'].unique()
```

```python
only_fives = ratings[ratings['rating'] == 5.0]
only_fives.groupby('movieId', as_index=False).agg({'rating': 'count'})\
    .sort_values('rating', ascending=False).head(10)
```

```python
# возьмем фильм с самым большим количеством оценок 5


film_id =only_fives.groupby('movieId', as_index=False).agg({'rating': 'count'})\
    .sort_values('rating', ascending=False).head(1)['movieId'].values[0]

film_id
```

```python
top_5_film = movies[movies['movieId'] == film_id]


top_5_film
```

```python
print(f"Как и ожидалось, самый популярный фильм - {top_5_film['title'].values[0]}")
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
power = pd.read_csv('1_home_work/task_2/power.csv')
transactions = pd.read_csv('1_home_work/task_2/transactions.csv')
```

```python
power.head()
```

```python
transactions.head()
```

```python
# power['country'].unique() # Latvia, Lithuania, Estonia
```

```python
pribaltik = power[(power['country'] == 'Latvia') | (power['country'] == 'Lithuania') | (power['country'] == 'Estonia')]
```

```python
pribaltik =pribaltik[(pribaltik['year'] >= 2005) & (pribaltik['year'] <= 2010)]
```

```python
pribaltik = pribaltik[(pribaltik['category'] == 4) | (pribaltik['category'] == 12) | (pribaltik['category'] == 21)]
```

```python
pribaltik['year'].unique(), pribaltik['country'].unique(), pribaltik['category'].unique()
```

```python
total_pribalt = pribaltik[pribaltik['quantity'] > 0]['quantity'].sum()
```

```python
print(f"Общее потребление стран Прибалтики с 2005 по 2010 гг. составило {total_pribalt}")
```
