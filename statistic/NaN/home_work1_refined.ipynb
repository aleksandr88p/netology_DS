    если посмотреть глазами, то выбросы могут быть в pulse, respiratory_rate. потому что там очень сильно значения минимального и максимального отличаются от среднего. И скорее всего они связаны между собой,
    типа большой пульс => чаще дышит. так что посчитаю выбросы для этих столбцов

```python
def remove_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_b = q1 - (1.5 * iqr)
    upper_b = q3 + (1.5 * iqr)
    return df[df[column].between(lower_b, upper_b, inclusive="both")]

remove_outliers_respiratory_rate = remove_outliers(df, 'respiratory_rate')
remove_outliers_rectal_temperature = remove_outliers(df, 'rectal_temperature')
remove_outliers_pulse = remove_outliers(df, 'pulse')
```

```python
def fill_pain(row):
    if pd.isna(row['pain']):  # Проверяем, пустое ли значение
        if row['outcome'] == 2:
            # Сравниваем с медианными значениями pulse
            if abs(row['pulse'] - 75) < abs(row['pulse'] - 84):
                return 4  # Ближе к 75
            else:
                return 5  # Ближе к 84
        elif row['surgery'] == 2:
            return 1  # Если surgery == 2, устанавливаем pain = 1
        elif row['temperature_of_extremities'] == 3:
            return 2  # Если temperature_of_extremities == 3, устанавливаем pain = 2
        elif row['temperature_of_extremities'] == 2:
            return 3  # Если temperature_of_extremities == 2, устанавливаем pain = 3
        else:
            # Условие для остальных случаев
            if row['pulse'] < 60:
                return 1
            elif row['pulse'] < 70:
                return 2
            elif row['pulse'] < 80:
                return 3
            else:
                return 4
    return row['pain']  # Если значение не пустое, оставляем как есть
```

```python
df2['pain'] = df2.apply(fill_pain, axis=1)
```

```python

import pandas as pd
import numpy as np

# Load the dataset
h_data = 'horse_data.csv'
columns = ['surgery?', 'Age', 'rectal temperature', 'pulse', 'respiratory rate',
           'temperature of extremities', 'pain', 'outcome', 'other_columns']
df = pd.read_csv(h_data, header=None, na_values='?', names=columns, usecols=range(8))

# Initial analysis of data
basic_stats = df.describe(include='all')

# Handling missing values based on column types
# For categorical data, fill missing values with mode (most frequent value)
categorical_columns = ['surgery?', 'Age', 'temperature of extremities', 'pain', 'outcome']
for column in categorical_columns:
    df[column] = df[column].fillna(df[column].mode()[0])

# For numerical data, analyze outliers and fill with median
numerical_columns = ['rectal temperature', 'pulse', 'respiratory rate']
for column in numerical_columns:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    # Remove outliers for median calculation
    filtered_data = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    df[column] = df[column].fillna(filtered_data[column].median())

# Save the cleaned dataset
cleaned_file = 'cleaned_horse_data.csv'
df.to_csv(cleaned_file, index=False)

# Verify the results
missing_values_summary = df.isnull().sum()
cleaned_stats = df.describe(include='all')
```
