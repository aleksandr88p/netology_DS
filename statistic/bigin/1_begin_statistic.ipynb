# Мода
    при подстчете вручную будет отличаться от того как это будет встроенной функцией пандас из за округления.
    то как округляю я и как пандас происходит по разному. И что бы результаты были одинаковые, надо заранее их
    округлить 

```python
# вручную
mode_counts = {}
for i in df['Weight'].round():
    if i not in mode_counts:
        mode_counts[i] = 1
    else:
        mode_counts[i] += 1
        
maxcount = 0
mode_value = None
for k, v in mode_counts.items():
    if maxcount < v:
        maxcount = v
        mode_value = k
        
print(f"значение моды: {mode_value}, частота {maxcount}")
```

#  средне квадратичное отклонение
    это корень дисперсии

```python
# вручную
def stdev(nums):
    diffs = 0
    avg = sum(nums) / len(nums) # mean
    for n in nums:
        # сумма квадратичный отклонений
        diffs += (n - avg) ** (2)
    # корень среднеквадратичного отклонения    
    return (diffs / len(nums) - 1) ** (0.5)

print("Height")
print(stdev(df['Height']))
print("Weight")
print(stdev(df['Weight']))
```

```python
print("Height")
print(df['Height'].std())
print(np.std(df['Height']))
print(np.std(df['Height'], ddof=1)) # дельта степеней свободу в нумпай по умолчанию 0
print("Weight")
print(df['Weight'].std())
print(np.std(df['Weight'], ddof=1)) # дельта степеней свободу в нумпай по умолчанию 0
```

# Дисперсия

```python
# вручную
def disp(nums):
    diffs = 0
    avg = sum(nums) / len(nums) # mean
    for n in nums:
        # сумма квадратичный отклонений
        diffs += (n - avg) ** (2)
    return (diffs / len(nums) - 1)

print(disp(df['Height']))
print(disp(df['Weight']))
```

```python
print("Height")
print(df['Height'].var())
print(np.var(df['Height']))
print(np.var(df['Height'], ddof=1)) # дельта степеней свободу в нумпай по умолчанию 0
print("Weight")
print(df['Weight'].var())
print(np.var(df['Weight'], ddof=1)) # дельта степеней свободу в нумпай по умолчанию 0
```

```python
import numpy as np

# trials - список из 1 млн. целых чисел от 0 до 999
# генерируется отдельно в тестовом файле
def monte_carlo(trials):
    # словарь, содержащий выигрышные комбинации и сумму выигрыша
    # случаи с * следует обработать в цикле с помощью
    # операции взятия остатка от деления (%)
    score = {
        999: 100,
        777: 200,
        555: 50,
        333: 15,
        111: 10
    }
    # сумма выигрыша      
    balance = 0

    for i in trials:
        if i in score:
            balance += score[i]
        elif i % 100 == 77:  # x77
            balance += 5
        elif i % 10 == 7:    # xx7
            balance += 3
        elif i % 100 == 0:   # x00
            balance += 2
        elif i % 10 == 0:    # xx0
            balance += 1
        

    res = balance / len(trials)
    return round(res, 5)


np.random.seed(42)
trials = np.random.randint(0, 1000, 1000000)
print(monte_carlo(trials))
```
