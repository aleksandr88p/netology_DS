Given a variable that contains a word consisting of Latin letters, write code that prints:

- the middle letter if the number of letters in the word is odd;
- the two middle letters if the number of letters is even.


```python
word = input("Enter a word: ")
letter_nedeed = len(word)//2
if len(word) % 2 == 0:
    print(word[letter_nedeed-1:letter_nedeed+1])
else:
    print(word[letter_nedeed])

length = len(word)
print(word[(length - 1) // 2 : (length + 2) // 2])
```

Write a program that sequentially asks the user for numbers (one at a time) and, after the first zero, outputs the sum of all previously entered numbers.


```python
total_sum = 0
while True:
    try:
        n = float(input())
        if n == 0:
            break
        total_sum += n
    except ValueError:
        print("Entering anything other than numbers is not allowed.")

if total_sum.is_integer():
    print(int(total_sum))
else:
    print(total_sum)
```

We are creating an MVP for a dating service, and we have a list of boys and girls. We propose a hypothesis: the best recommendations can be made by simply sorting the names alphabetically and matching people with the same index after sorting! However, we won’t match anyone if there’s a chance that someone might be left without a partner.


```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']


if len(boys) == len(girls):
    boys.sort()
    girls.sort()
    print("Идеальные пары:")
    for boy, girl in zip(boys, girls):
        print(f"{boy} и {girl}")
else:
    print("Внимание, кто-то может остаться без пары!")


boys_sorted, girls_sorted = sorted(boys), sorted(girls)
print("Идеальные пары:\n"+"\n".join(f"{boy} и {girl}" for boy, girl in zip(boys_sorted, girls_sorted)) if len(boys_sorted) == len(girls_sorted) else "Внимание, кто-то может остаться без пары!")

```

We have a list containing information about the average daily temperature in Fahrenheit for a random period by country (the data structure is provided in the example). You need to write code that calculates the average temperature over the period in Celsius (!) for each country.

```python


countries_temperature = [
    ['Таиланд', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Германия', [57.2, 55.4, 59, 59, 53.6]],
    ['Россия', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Польша', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

print("Средняя температура в странах:")
for country in countries_temperature:
    avg_temp_f = sum(country[1]) / len(country[1])
    celsius_temp = (avg_temp_f - 32) * 5 / 9
    print(f"{country[0]} - {celsius_temp:.1f} C")
```

There is a list of vehicle registration numbers. You need to write code that checks each number for validity (1 letter, 3 digits, 2 letters, 2-3 digits). Note that not all Cyrillic alphabet letters are used in vehicle registration numbers.

```python
import re
car_ids = ['А222ВС96', 'АБ22ВВ193']

letters = ['А', 'В', 'Е', 'К', 'М', 'Н', 'О', 'Р', 'С', 'Т', 'У', 'Х']

for plate in car_ids:
    if (
        len(plate) in [8, 9] and
        plate[0] in letters and
        plate[4] in letters and
        plate[5] in letters and
        plate[1:4].isdigit() and
        plate[6:].isdigit()
    ):
        region = plate[6:]
        print(f"Номер {plate[:6]} валиден. Регион: {region}")
    else:
        print(f"Номер {plate} не валиден")


#letter_for_regex = ''.join(letters)
#print(letter_for_regex)
pattern = re.compile(r'^[АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2}\d{2,3}$')
for plate in car_ids:
    match = pattern.match(plate)
    if match:
        region = plate[-2:] if len(plate) == 8 else plate[-3:]
        print(f"Номер {plate[:-len(region)]} валиден. Регион: {region}")
    else:
        print(f"Номер {plate} не валиден")
```

Дан поток логов по количеству просмотренных страниц для каждого пользователя (пользователь,дата;просмотры). Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей. И тут регулярные выражения облегчат немного реализацию.

```python
stream = [
    'user4,2021-01-01;3',
    'user3,2022-01-07;4',
    'user2,2022-03-29;1',
    'user1,2020-04-04;13',
    'user2,2022-01-05;7',
    'user1,2021-06-14;4',
    'user3,2022-07-02;10',
    'user4,2021-03-21;19',
    'user4,2022-03-22;4',
    'user4,2022-04-22;8',
    'user4,2021-05-03;9',
    'user4,2022-05-11;11'
]

views_by_user = {}
for log in stream:
    user_id, date_views = log.split(',')
    date, views = date_views.split(';')
    views = int(views)

    if user_id in views_by_user:
        views_by_user[user_id] += views
    else:
      views_by_user[user_id] = views

total_views = sum(views_by_user.values())
print(f"Среднее количество просмотров на уникального пользователя: {total_views / len(views_by_user):.2f}")

```
